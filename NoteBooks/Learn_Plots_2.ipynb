```julia
using Plots
```

```julia
theme(:dark)
```

```julia
plot(x -> sin(x),
    label = "Sin(x)",
    xlim = (0,2π), ylim = (-1.10,1.10),
    xticks = ((0:π/2:2π),["0","π/2","π","3π/2","2π"]), tickfontsize=10,
    dpi=500)

plot!(x -> cos(x), label = "Cos(x)",dpi = 500 )
```

### subplot

```julia
p1 = plot(x -> sin(x), 
    label = "Sin(x)",
    xlim = (0,2π), 
    xticks = ((0:π/2:2π),["0","π/2","π","3π/2","2π"]), 
    ylim = (-1.1,1.1),
    yticks = ((-1.1:0.5:1.1),["-1.0","-0.5","0.0","0.5","1.0"]),
    color = :magenta,
    dpi = 350)
p2 = plot(x -> cos(x), 
    label = "cos(x)",
    xlim = (0,2π), 
    xticks = ((0:π/2:2π),["0","π/2","π","3π/2","2π"]), 
    ylim = (-1.1,1.1),
    yticks = ((-1.1:0.5:1.1),["-1.0","-0.5","0.0","0.5","1.0"]),
    color = :lightseagreen, 
    dpi = 350)
plot(p1, p2,p2,p1, layout = (2,2))
# plot!(p2, p1, layout = (2,1))
```

### annotation

```julia

p1 = plot(x -> sin(x), 
    label = "Sin(x)",
    xlim = (0,2π), 
    xticks = ((0:π/2:2π),["0","π/2","π","3π/2","2π"]), 
    ylim = (-1.1,1.1),
    yticks = ((-1.1:0.5:1.1),["-1.0","-0.5","0.0","0.5","1.0"]),
    color = :magenta,
    dpi = 350)
p2 = plot(x -> cos(x), 
    label = "cos(x)",
    xlim = (0,2π), 
    xticks = ((0:π/2:2π),["0","π/2","π","3π/2","2π"]), 
    ylim = (-1.1,1.1),
    yticks = ((-1.1:0.5:1.1),["-1.0","-0.5","0.0","0.5","1.0"]),
    color = :lightseagreen, 
    dpi = 350)
# plot(p1, p2,p2,p1, layout = (2,2))
plot(p1)
annotate!(π/2,1.05, ("Local Max",10,:gold,:center))
scatter!([π/2],[1.0],color=:red,label="extrema")

annotate!(3π/2,-1.05,("Local Min",10,:gold,:center))
scatter!([3π/2],[-1.0],color=:royalblue,label="extrema")
```
